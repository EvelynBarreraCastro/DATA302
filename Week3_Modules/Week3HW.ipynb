## Problem 3

Scroll through the packages on the pypi.org index and find one that looks interesting to you. In the cell below, describe the package and what it is used for. 

```python
# Program below is used to play Star Wars
import socket
import sys

def play_star_wars():
    print("Connecting to a galaxy far, far away... (Press Ctrl+C to quit)")
    
    # Connect directly to the famous ASCII Star Wars server
    host = "towel.blinkenlights.nl"
    port = 23
    
    try:
        # Open a low-level network connection
        s = socket.create_connection((host, port))
        
        while True:
            # Read streaming text frames from the server
            data = s.recv(1024)
            if not data:
                break
            # Print the text animation frame to your terminal screen
            sys.stdout.write(data.decode('utf-8', errors='ignore'))
            sys.stdout.flush()
            
    except KeyboardInterrupt:
        print("\nMay the Force be with you!")
    except Exception as e:
        print(f"\nCould not connect: {e}")
        print("Note: Some modern school or home networks block Port 23 traffic.")

if __name__ == "__main__":
    play_star_wars()
```
