```python
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline
```

Our main focus this week will be line and scatter plots, each of which starts with a list, tuple, or Series of data to display. As we will see, the plotting functions take many parameters that change the final product. 

```python
plotting_list = [1,2,6,-12,5,3]
```

```python
plt.plot(plotting_list)
```

A simple line plot like the one above takes as input a single vector and plots those values as the y-values against integer x-values starting at 0. If it is provided two vectors, the first is taken as the x-values and the second is taken to be the y-values.  We can change the color and many other features of the plot - the next example has some example parameters, take some time to see what happens as you adjust them. 

```python
x_positions = [1,-3,6,3,12]
y_positions = [3,5,8,-4,2]
```

```python
plt.plot(x_positions,y_positions, linewidth=5, color = 'orange', linestyle = 'dashed', marker = 'o', markersize = 35)
```

Once we've settled on the features of the plot we want, the next step is to add appropriate labels and other text elements. After defining the plot, we will use functions like `plt.title` and `plt.xlabel` to add annotations. We can also control the labels and positions of the ticks with `xticks` and `yticks`. 

```python
plt.plot(x_positions,y_positions, linewidth=5, color = 'hotpink', linestyle = 'dashed', marker = '*', markersize = 30)
plt.title("Very Important Data")
plt.xlabel("X values")
plt.ylabel("Y values")
```

```python
plt.plot(x_positions,y_positions, linewidth=5, color = 'hotpink', linestyle = 'dashed', marker = '*', markersize = 30)
plt.title("Very Important Data")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.xticks([-5,0,5,10,15],['Super small', 'Very small', 'Kindof small','Large', 'Giant'],rotation=45)
plt.yticks(range(-5,10),range(-5,10),fontsize=20)
```

We can construct a scatterplot using the `plt.plot` function by leaving out the linestyle. However, there is also a separate scatterplot function (`plt.scatter`) that has some slightly different parameter designations.  

```python
plt.scatter(x_positions,y_positions,  color = 'hotpink',  marker = 'o', s=200)
plt.title("Very Important Data")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.xticks([-5,0,5,10,15],['Super small', 'Very small', 'Kindof small','Large', 'Giant'],rotation=45)
plt.yticks(range(-5,10),range(-5,10),fontsize=20)
```

```python
first_point_colors = ['blue','red','orange','purple','gray']
second_point_colors = [1,4,5,2,4]
```

```python
plt.scatter(x_positions,y_positions,  c = first_point_colors,  marker = '+', s=200)
plt.title("Very Important Data")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.xticks([-5,0,5,10,15],['Super small', 'Very small', 'Kindof small','Large', 'Giant'],rotation=45)
plt.yticks(range(-5,10),range(-5,10),fontsize=20)
```

```python
plt.scatter(x_positions,y_positions,  c = second_point_colors,  marker = '+', s=200, cmap ="coolwarm")
plt.title("Very Important Data")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.xticks([-5,0,5,10,15],['Super small', 'Very small', 'Kindof small','Large', 'Giant'],rotation=45)
plt.yticks(range(-5,10),range(-5,10),fontsize=20)
```
